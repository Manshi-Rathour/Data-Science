```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Day = [1,2,3,4,5,6,7]
Temperature = [30, 32, 26, 28, 24, 26, 30]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(Day, Temperature, color = 'green', linewidth = 4, linestyle = 'dotted')
```

```python
plt.plot(Day, Temperature, color = 'red', marker = '+', linestyle = '', markersize = 20)
```

```python
plt.plot(Day, Temperature, color = 'purple', alpha = 0.5)
```

```python
day = [1,2,3,4,5,6,7]
max_t = [50, 45, 47, 48, 52, 50, 54]
min_t = [38, 40, 41, 44, 36, 32, 34]
avg_t = [45, 47, 43, 48, 42, 44, 40]
```

```python
plt.plot(day, max_t, label='Max')
plt.plot(day, min_t, label='Min')
plt.plot(day, avg_t, label='Avg')
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.grid()
plt.legend()
```

```python
subject = ['maths', 'physics', 'mechanics', 'cs', 'tc']
student1_marks = [95, 90, 92, 98, 86]
student2_marks = [92, 96, 90, 93, 91]
```

```python
plt.plot(subject, student1_marks, label = 'student1', linewidth = 4)
plt.plot(subject, student2_marks, label = 'student2', linewidth = 3)
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title("Student's marks in different subjects")
plt.legend()
plt.grid()
```
