```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
company = ['GOOGL', 'AMAZN', 'MRSFT', 'FB']
revenue = [95, 136, 89, 36]
profit = [40, 5, 12, 17]
```

```python
xpose = np.arange(len(company))
xpose
```

```python
plt.xticks(xpose, company)
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpose, revenue)
```

```python
plt.xticks(xpose, company)
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpose, revenue, label='Revenue')
plt.bar(xpose, profit, label='Profit')
plt.legend()
```

```python
plt.xticks(xpose, company)
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpose-0.1, revenue, width = 0.4, label = 'Revenue')
plt.bar(xpose+0.1, profit, width = 0.4, label = 'Profit')
plt.legend()
```
