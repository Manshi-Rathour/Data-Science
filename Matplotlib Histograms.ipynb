```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [100, 95, 130, 88, 145, 98, 89, 105, 103, 117, 78, 139, 82, 150]
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95)
plt.xlabel('Sugar range')
plt.ylabel('Total no. of patient')
plt.title('Blood Sugar Analysis')
```

```python
blood_sugar_men = [100, 95, 130, 88, 145, 98, 89, 105, 103, 117, 78, 139, 82, 150]
blood_sugar_women = [69, 77, 89, 80, 112, 130, 117, 89, 95, 86, 115, 145, 136, 129]
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['red', 'pink'],
         label = ['Men', 'Women'])
plt.xlabel('Sugar range')
plt.ylabel('Total no. of patient')
plt.title('Blood Sugar Analysis')
plt.legend()
```
