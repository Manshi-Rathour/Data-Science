```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weather_data.csv')
df
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
new_df = df.fillna(method = "ffill")
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.replace([30, 10], np.NaN)
new_df
```

```python
new_df = df.replace({
    'temperature':29,
    'windspeed':6,
    'event':'snow'
}, np.NaN)
new_df
```

```python
new_df = df.replace({
    29 : np.NaN,
    'sunny':'no event'
})
new_df
```
